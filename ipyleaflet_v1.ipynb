```python
from ipyleaflet import Map, basemaps, basemap_to_tiles
```

```python
m = Map(center=(52.204793, 360.121558), zoom=9)

dark_matter_layer = basemap_to_tiles(basemaps.CartoDB.DarkMatter)
m.add_layer(dark_matter_layer)
m
```

```python
m = Map(center=(55.991468, 36.899394), zoom=13)

matter_layer = basemap_to_tiles(basemaps.CartoDB) #(basemaps.CartoDB.DarkMatter)
m.add_layer(matter_layer)
m
```

```python
from ipyleaflet import Map, MagnifyingGlass, basemaps, basemap_to_tiles

# m = Map(center=(0, 0), zoom=2)
m = Map(center=(55.991468, 36.899394), zoom=4)

topo_layer = basemap_to_tiles(basemaps.OpenTopoMap)
magnifying_glass = MagnifyingGlass(layers=[topo_layer])

m.add_layer(magnifying_glass)

m
```

```python
from ipyleaflet import Map, Marker

center = (55.991468, 36.899394)

#52.204793, 360.121558
#55.991468, 36.899394

m = Map(center=center, zoom=15)

marker = Marker(location=center, draggable=False)
m.add_layer(marker);

m
```

```python
from ipyleaflet import Marker, Icon, Map

center = (55.991468, 36.899394)

m = Map(center=center, zoom=10)
icon = Icon(icon_url='https://leafletjs.com/examples/custom-icons/leaf-green.png', icon_size=[38, 95], icon_anchor=[22,94])
mark = Marker(location=center, icon=icon, rotation_angle=90, rotation_origin='22px 94px')
m.add_layer(mark);

m
```
